```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('revelleit.db')
```

```python
# Tableau de tous les desktops classés par score
all_desktops = pd.read_sql_query("""
    SELECT DISTINCT Marque, Modele, Departement, 
           GWP, ADPe, WU, ADPf, TPE, WEEE, Score_GWP, Score_ADPe, Score_WU, Score_ADPf, Score_TPE, Score_WEEE,Green_IT_Score
    FROM impacts_calcules
    ORDER BY Green_IT_Score DESC
""", conn)
# éliminer les éventuels doublons résiduels
all_desktops = all_desktops.drop_duplicates()
print("Tous les équipements classés par score :")
all_desktops
```

```python
top_5 = pd.read_sql_query("""
    SELECT DISTINCT Marque, Modele, 
           GWP, ADPe, WU, ADPf, TPE, WEEE, Score_GWP, Score_ADPe, Score_WU, Score_ADPf, Score_TPE, Score_WEEE,Green_IT_Score
    FROM impacts_calcules
    ORDER BY Green_IT_Score DESC
    LIMIT 5
""", conn)
# cas où plusieurs lignes identiques passeraient encore
top_5 = top_5.drop_duplicates()

print("Top 5 des équipements avec le meilleur score")
top_5
```

```python
# Calcul du score moyen par département
dept_scores = pd.read_sql_query("""
    SELECT Departement, AVG(Green_IT_Score) AS Score_Moyen
    FROM impacts_calcules
    GROUP BY Departement
    ORDER BY Score_Moyen DESC
""", conn)

# score moyen de chaque KPI par département
dept_kpi = pd.read_sql_query("""
    SELECT Departement,
           AVG(Score_GWP) AS Avg_Score_GWP,
           AVG(Score_ADPe) AS Avg_Score_ADPe,
           AVG(Score_WU)  AS Avg_Score_WU,
           AVG(Score_ADPf) AS Avg_Score_ADPf,
           AVG(Score_TPE) AS Avg_Score_TPE,
           AVG(Score_WEEE) AS Avg_Score_WEEE
    FROM impacts_calcules
    GROUP BY Departement
""", conn)

# fusion pour affichage conjoint
dept_combined = dept_scores.merge(dept_kpi, on='Departement')
print("Score moyen global et par KPI par département :")
dept_combined


```

```python
# statistiques brutes par département (moyennes et max pour chaque KPI)
print("\nStatistiques brutes par département :")
dept_raw = pd.read_sql_query("""
    SELECT Departement,
           AVG(GWP) AS Avg_GWP, AVG(TPE) AS Avg_TPE,
           AVG(ADPe) AS Avg_ADPe, AVG(WU) AS Avg_WU,
           AVG(ADPf) AS Avg_ADPf, AVG(WEEE) AS Avg_WEEE,
           MAX(GWP) AS Max_GWP, MAX(TPE) AS Max_TPE
    FROM impacts_calcules
    GROUP BY Departement
""", conn)
dept_raw
```

```python
# Score moyen par KPI par département + score moyen global par département
kpi_dept = pd.read_sql_query("""
    SELECT Departement,
           AVG(Score_GWP) AS Moy_GWP,
           AVG(Score_ADPe) AS Moy_ADPe,
           AVG(Score_WU)  AS Moy_WU,
           AVG(Score_ADPf) AS Moy_ADPf,
           AVG(Score_TPE) AS Moy_TPE,
           AVG(Score_WEEE) AS Moy_WEEE,
           AVG(Green_IT_Score) AS Moy_Score_Global
    FROM impacts_calcules
    GROUP BY Departement
    ORDER BY Moy_Score_Global DESC
""", conn)
print("Score moyen global et par KPI par département :")
kpi_dept
```

```python
# recherche des PC les plus consommateurs par KPI
print("\nPC ayant la valeur la plus élevée par KPI :")
desktop_stats = pd.read_sql_query("""
    SELECT Marque, Modele,
           GWP, ADPe, WU, ADPf, TPE, WEEE
    FROM impacts_calcules
""", conn)
# utilisation de pandas pour extraire les maxima
for kpi in ['GWP','ADPe','WU','ADPf','TPE','WEEE']:
    idx = desktop_stats[kpi].idxmax()
    print(f"{kpi} max = {desktop_stats.loc[idx,'Marque']} {desktop_stats.loc[idx,'Modele']} ({desktop_stats.loc[idx,kpi]})")
```

```python
# afficher le département avec la moyenne la plus élevée pour chaque KPI
print("\nDépartement leader par KPI (moyenne) :")
dept_leaders = pd.read_sql_query("""
    SELECT Departement,
           AVG(GWP) AS Avg_GWP, AVG(ADPe) AS Avg_ADPe, AVG(WU) AS Avg_WU,
           AVG(ADPf) AS Avg_ADPf, AVG(TPE) AS Avg_TPE, AVG(WEEE) AS Avg_WEEE
    FROM impacts_calcules
    GROUP BY Departement
""", conn)
for k in ['Avg_GWP','Avg_ADPe','Avg_WU','Avg_ADPf','Avg_TPE','Avg_WEEE']:
    idx = dept_leaders[k].idxmax()
    print(f"{k} leader = {dept_leaders.loc[idx,'Departement']} ({dept_leaders.loc[idx,k]})")
```
